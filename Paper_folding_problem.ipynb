problem #1 Implementation using exponentiation arithmetic operators

```python
# Code to calculate the thickness when the paper is folded 43 times
THICKNESS = 0.00008

folded_thickness = THICKNESS *2** 43

print("Thickness: {} meters".format(folded_thickness))
```

problem #2 Unit Conversion

```python
# Convert meters to kilometers and display with two decimal places
print("Thickness: {: .2f} kilometers".format(THICKNESS / 1000))

# Let's check if the folded paper arrives in the moon
distance_to_moon =384400
if THICKNESS/1000 > distance_to_moon:
    print("Folded paper exceeds distance to moon")
```

problem #3 Create using a for statement

```python
THICKNESS = 0.00008
folding_times = 43
for i in range(folding_times):
    THICKNESS = THICKNESS*2
print("Thickness: {} meters".format(THICKNESS))
```

problem #4  Comparison of calculation time

```python
import time
```

```python

start = time.time()
#####
THICKNESS = 0.00008

folded_thickness = THICKNESS *2** 43

print("Thickness: {} meters".format(folded_thickness))
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

problem #5 Saving to a list

```python
folded_thichness = [] #create list
folded_thichness.append(THICKNESS) #initial value
for i in range(folding_times):
    THICKNESS = THICKNESS * 2
    folded_thichness.append(THICKNESS)
 
len(folded_thichness)
```

problem #6 Displaying a line graph

```python
import matplotlib.pyplot as plt

# Display the graph. Title and axis label name.
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thichness)  
plt.show()

# Brief explanation: Thickness increases 2 times more for each fold.
```

Problem #7 Customizing graphs

```python
import matplotlib.pyplot as plt

# Display the graph. Title and axis label name.
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.plot(folded_thichness, linestyle=':', color='green', linewidth=3)
plt.show()
```

```python
# Display the graph. Title and axis label name.
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.plot(folded_thichness, color='red', linewidth=2)
plt.show()
```

```python
# Display the graph. Title and axis label name.
plt.title("thickness of folded paper")
plt.xlabel("number of folds", fontsize =20, color="blue")
plt.ylabel("thickness [m]", fontsize =20, color="green")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.grid(True)
plt.plot(folded_thichness, color='brown', linewidth=2)
plt.show()
```
